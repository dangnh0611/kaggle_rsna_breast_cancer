```python
import numpy as np
import cv2
import os
import sys
import random
import time
```

```python
DIR1 = '/home/dangnh36/projects/.comp/rsna/datasets/exports/uint16_png'
DIR2 = '/raid/.comp/export_uint16_png_v2'
SAVE_DIR = '/raid/.comp/temp'
os.makedirs(SAVE_DIR, exist_ok = True)
```

```python
names = os.listdir(DIR2)
names = random.choices(names, k = 100)
for name in names:
    path1 = os.path.join(DIR1, name)
    path2 = os.path.join(DIR2, name)
    t0 = time.time()
    img1 = cv2.imread(path1, cv2.IMREAD_ANYDEPTH)
    img2 = cv2.imread(path2, cv2.IMREAD_ANYDEPTH)
    t1 = time.time()
    assert img1.dtype == np.uint16
    assert img2.dtype == np.uint16
    
    diff = np.max(np.abs(img1 - img2))
    if diff != 0:
        print('invert!')
        img2 = img2.max() - img2
        diff = np.max(np.abs(img1 - img2))
    print(name, diff, round(t1 - t0, 2), ' | ', img1.min(), img1.max(), img2.min(), img2.max() )
    assert diff == 0
    save_np_path = os.path.join(SAVE_DIR, name.replace('.png', '.npy'))
    t2 = time.time()
    np.save(save_np_path, img1)
    t3 = time.time()
    print('save np:', t3 - t2)
    
```

```python
takes = []
for name in os.listdir(SAVE_DIR):
    t0 = time.time()
    img = np.load(os.path.join(SAVE_DIR, name))
    print(img.shape)
    t1 = time.time()
    take = t1 - t0
    takes.append(take)
    print('Take:', take)

takes = np.array(takes)
print(takes.mean(), takes.std())
```
